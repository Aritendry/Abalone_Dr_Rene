```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('../data/abalone.csv')
data.head()
```

```python
data.columns
```

```python
df = data.drop("Unnamed: 0", axis=1)
```

```python
df.describe()
```

```python
df.columns
```

```python
X = df.drop("Rings", axis=1)
y = df["Rings"]
```

```python
std = StandardScaler()
x_std = std.fit_transform(X)
y_std = std.fit_transform(y.values.reshape(-1, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_std, y_std, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
rl = LinearRegression()
rl.fit(X_train, y_train)
```

```python
rd = Ridge()
rd.fit(X_train, y_train)
```

```python
y_predrd = rd.predict(X_test)
```

```python
y_pred = rl.predict(X_test)
```

```python
print("mae : ", mean_absolute_error(y_test, y_predrd))
print("r2 : ", r2_score(y_test, y_predrd))
```

```python
print("mae : ", mean_absolute_error(y_test, y_pred))
print("r2 : ", r2_score(y_test, y_pred))
```

```python
std.inverse_transform(y_predrd.reshape(-1, 1)).ndim
```

```python
y_test.ndim
```

```python
std.inverse_transform(y_pred).ndim
```

```python
y_pred_tab = pd.DataFrame({"y_test": y_test, "y_pred": std.inverse_transform(y_pred), "y_predrd": std.inverse_transform(y_predrd.reshape(-1, 1))})
y_pred_tab.head()
```

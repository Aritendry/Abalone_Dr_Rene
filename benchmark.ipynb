```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_abalone = pd.read_csv('Dr_rene/abalone.data' , header=None)
"""
Lien : https://en.wikipedia.org/wiki/Abalone
Abalone est un molusque marin il est appelé escargot de mer

"""
```

```python
df_abalone.head(10)
```

```python
df_abalone.info()
```

```python
df_abalone.columns = ['Sex','Length','Diameter','Height','Whole_weight','Shucked_weight','Viscera_weight','Shell_weight','Rings']
```

```python
df_abalone.head()
```

```python
df_abalone.isnull().sum()
```

```python
print(f"Valeurs unique de Sex : {df_abalone['Sex'].unique()}")
```

```python
mapping = {'I':2 , 'F':1 , 'M':0}
df_abalone['Sex'] = df_abalone['Sex'].replace(mapping)
```

```python
df_abalone.head(10)
```

```python
sns.pairplot(df_abalone)
plt.show()
```

```python
corr_matrix = df_abalone.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix , annot=True , cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(x='Sex' , data=df_abalone)
plt.title('Histogramme de Sex')
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.histplot(df_abalone['Rings'], kde = True)
plt.title("Histogram of Rings")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'Rings', data = df_abalone, hue = 'Sex', palette = 'Set2')
plt.legend(loc = 'lower right')
plt.title('Boxplot of Rings by Sex')
plt.xlabel("Rings")
plt.ylabel("Sex")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'Whole_weight', data = df_abalone, hue = 'Sex', palette = 'Set2')
plt.legend(loc = 'lower right')
plt.title('Boxplot of Weight by Sex')
plt.xlabel("Weight")
plt.ylabel("Sex")
plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.scatterplot(x = 'Whole_weight', y = 'Diameter', hue = 'Sex', data = df_abalone, palette = 'Set1')
plt.legend(loc = 'lower right')
plt.title("Weight Versus Diameter")
plt.xlabel("Weight")
plt.ylabel("Diameter")
plt.show()
```
